## Working with pivot tables

Q1. **Pivot** the weather by keeping value as "Rainfall", **index** as "MinTemp" and **columns** as "MaxTemp".

Q2. Slice **first five index** 

Q3. Calculate summary stats across index and columns separately, i.e (mean)

```python
#1
## Start the code here
weatherpivot=weather.pivot_table(values='Rainfall',index='MinTemp',columns='MaxTemp')
weatherpivot
## End here
```

```python
#2
## Start the code here
weatherpivot.iloc[0:5]
## End here
```

```python
#3
## Start the code here
print('Mean for Columns')
print('')
print(weatherpivot.mean(axis=0))
print('---------------------------')
print('')
print('Mean for Rows')
print('')
print(weatherpivot.mean(axis=1))
## End here
```

# Creating & Visualizing Dataframes

## Visualizing your data

Q1. Plot histogram of **"Rainfall"** keeping bins as 25 and show it.

**Bar plots**

Q2. Assign **"Rainfall"** & **"RainToday"** to df and perform bar plot.

**Line plot**

Q3. Assign **"RainToday"** and **"WindGustSpeed"** to df1 and perform line plot

**Layering plots** 

Q4. Layer the plots keeping the dataframe as **"RainTomorrow" & "Rainfall"** and also add a **legend("Yes","No")**

Q5. Add Transperancy , i.e alpha of 0.5

```python
# import matplotlib and pandas 
import pandas as pd
import matplotlib.pyplot as plt
# read the weather data
weather=pd.read_csv('weather.csv')
weather
```

```python
#1
## Start the code here
fig,ax=plt.subplots()
ax.hist(weather['Rainfall'],bins=25)
ax.set_xlabel('Rainfall')
ax.set_ylabel('# of Observations')
plt.show()
## End here
```

```python
#2
## Start the code here
#For barplot x axis should be categorical variable so took RainToday as X axis variable.
df=weather[['Rainfall','RainToday']]
fix,ax=plt.subplots()
ax.bar(df['RainToday'],df['Rainfall'])
ax.set_xlabel('Rain Today')
ax.set_ylabel('Rain Fall')
ax.set_title('RainToday  VS  RainFall')
plt.show()
## End here
```

```python
#3
## Start the code here
df1=weather[['RainToday','WindGustSpeed']]
fig,ax=plt.subplots()
ax.plot(df1['RainToday'],df1['WindGustSpeed'])
ax.set_xlabel('Rain Today')
ax.set_ylabel('WindGustSpeed')
ax.set_title('RainToday  VS  WindGustSpeed')
plt.show()
## End here
```

```python
#4
## Start the code here
dataframe=weather[['RainTomorrow','Rainfall']]
fig,ax=plt.subplots()
ax.plot(dataframe['RainTomorrow'],label='Yes')
ax.plot(dataframe['Rainfall'],label='No')
ax.legend()
plt.show()
## End here
```

```python
#5
## Start the code here
dataframe=weather[['RainTomorrow','Rainfall']]
fig,ax=plt.subplots()
ax.plot(dataframe['RainTomorrow'],label='Yes',alpha=0.5)
ax.plot(dataframe['Rainfall'],label='No',alpha=0.5)
ax.legend()
plt.show()
## End here
```
